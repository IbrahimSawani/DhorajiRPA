```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

Link = 'https://eurasiaonline.ie/lentils-oil/?sort=featured&page=1'

HTML = urlopen(Link)
HTML = HTML.read()

bs = BeautifulSoup(HTML, 'html.parser')
Product_Title = bs.findAll('h4', {'class' : 'card-title'})
Product_Price = bs.findAll('div', {'class' : 'card-text'})

for i in range(0, len(Product_Title)):
    Name = Product_Title[i].text
    Price = Product_Price[i].text
        
    print(Name.strip())
    print(Price.strip())    

Next_Page =  bs.find('li', {'class' : 'pagination-item pagination-item--next'})
Link = Next_Page.a.attrs['href']
    
while True:    
    
    Next_Page_html = urlopen(Link)
    Next_Page_html = Next_Page_html.read()
    bs = BeautifulSoup(Next_Page_html, 'html.parser')
    
    Product_Title = bs.findAll('h4', {'class' : 'card-title'})
    Product_Price = bs.findAll('div', {'class' : 'card-text'})
    
    for i in range(0, len(Product_Title)):
        Name = Product_Title[i].text
        Price = Product_Price[i].text
        
        print(Name.strip())
        print(Price.strip())
    
    
#     for i in Product_Title:
#         print(i.text)
#     for j in Product_Price:    
#         print(j.text)
    try:
        Next_Page =  bs.find('li', {'class' : 'pagination-item pagination-item--next'})
        Link = Next_Page.a.attrs['href']
    except:
        break
```
